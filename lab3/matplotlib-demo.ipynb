```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x, y, z = df.Height, df.Latitude, df.Longitude
```

```python
plt.scatter(x, y)
plt.savefig('scottish scatter plot.png')
```

```python
from scipy.stats import linregress
```

```python
stats = linregress(x, y)
m = stats.slope
b = stats.intercept
```

```python
plt.scatter(x, y)
plt.plot(x, m*x+b, 'red')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x, y, marker='o')
plt.plot(x, m*x+b, 'red', linewidth=3)
plt.xlabel('Height', fontsize=10)
plt.ylabel('Latitude', fontsize=10)
plt.xticks(fontsize=80)
plt.yticks(fontsize=80)
```

```python
def fahrenheit2celsius(temp):
    """Returns temperature in Celsius. """
    return (5. / 9.) * (temp - 32)
# TEMPURATURE AT EACH HOUR THROUGHOUT THE DAY (In Fahrenheit)
temperature  =  [100,  102,  106,  105,  90,  85,  85,  89,  100,  102,  103,  108,  100,  102,
106,  105,  90,  85,  85,  89,  100,  102,  103,  108]
```

```python
fig, ax_f = plt.subplots() 
ax_c = ax_f.twinx()

#plot our data:
ax_f.plot(temperature)
ax_f.set_xlim(0, 24) #x-axis shows 24 hours

# set the axis limits:
y1,   y2   =    ax_f.get_ylim() 
ax_c.set_ylim(fahrenheit2celsius(y1), fahrenheit2celsius(y2)) 
ax_c.figure.canvas.draw()

#  change  some  axis labels
ax_f.set_title('Two  scales:  Fahrenheit  and  Celsius') 
ax_f.set_ylabel('Fahrenheit') 
ax_c.set_ylabel('Celsius')
ax_f.set_xlabel('Time  (hour  of  day)')
```

```python
earnings = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/school_earnings.csv')
```

```python
earnings.head()
```

```python
plt.style.use("seaborn-pastel")
fig,ax=plt.subplots()
ax = plt.gca()
ax.set_facecolor('black')
ax.hist(earnings.Women,bins=25,density=True,histtype="stepfilled",alpha=0.8,label="women")
ax.hist(earnings.Men,bins=25,density=True,histtype="stepfilled",alpha=0.8,label="men")
ax.legend(prop={'size':10})
ax.set_ylabel("salary")
ax.set_xlabel("school")
```

```python
index = np.arange(len(earnings))
fig, ax = plt.subplots(figsize=(20,20)) 
bar_width = 0.35
opacity = 0.6

rects1  =  ax.barh(index,  earnings.Women, bar_width, alpha=opacity,  color='b', label='Women',)

# Note, shift the x postion also with the bar width so that the bar
 


# appears next to the previous oneś
rects2  =  ax.barh(index  +  bar_width,  earnings.Men,  bar_width, alpha=opacity, color='r', label='Men')

ax.set_xlabel('School') 
ax.set_ylabel('Earning')
ax.set_title('Earning gaps')
ax.set_yticks(index  +  bar_width  /  2) 
ax.set_yticklabels(earnings.School,rotation="vertical") 
ax.legend()
```
